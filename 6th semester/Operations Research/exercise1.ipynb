```python
import math
import numpy as np 

c = 0.8

def f(x):
    return 2*x[0] + x[1]**2 - math.log(x[0] + x[1] +1)

def gradf(x):  
    x1 = x[0]
    x2 = x[1]
    return np.array([ 2 - (1 / (x1+ x2 + 1)) , 2*x2 - (1 / (x1 + x2 + 1)) ])
```

```python
#ARMIJO-GOLDSTEIN
def armijo_goldstein(xn_1, xn):
  #print("x_n1 " , xn_1 , "xn" , xn)
  return f(xn_1) <= f(xn) + np.dot(c*(xn_1-xn),gradf(xn))
```

```python
def isDomain(xn):
  return xn[0] +xn[1] + 1 >0
```

```python
#Backtracking
def backtracking(xn, d):
  steps_back = 0
  t = 1
  xn_1 = xn - t*d
  while(True):
    steps_back = steps_back +1
    if(isDomain(xn_1) and armijo_goldstein(xn_1, xn)):
      return t , steps_back
    t = t/2
    xn_1 = xn - t*d
```

```python
def norm2(x):
    """
    Returns the euclidean norm of the vector x
    """
    x = np.array(x)
    return math.sqrt(x.dot(x))
```

```python
#gradient descent
xn = [3, 3]
steps = 0
tol = 10**(-7)
error = [100, 100]
while(True):
  d = gradf(xn)
  t, steps_back = backtracking(xn, d)
  steps = steps + steps_back
  xn_1 = xn - t*d
  error = xn_1 - xn
  if(norm2(error) <= tol):
    break
  print("X is " + str(xn) + "f is " + str(f(xn)) + "grad is " + str(gradf(xn)) + "direction is " , d , "t is " , t)

  xn = xn_1
  steps = steps + 1 
  
print(steps)
print(xn_1)
```
