```python
from pprint import pprint
def mult(A , B):
    """
    Matrix multiplication
    """
    result = []
    #For every line in A
    for i in range(len(A)):
        rowi = []
        #For every column in B
        for j in range(len(B[0])):
            tmp = 0
            for k in range(len(A[0])):
                tmp = tmp + A[i][k] * B[k][j]
            rowi.append(tmp)
        result.append(rowi)
    return result

A = [
    [-0.00000  , 1.00000] ,
   [0.50000  , 0.00000]
]
B = [
    [ 0  , 2] ,
    [1 , 0]
]
pprint(mult(A , B))
            
        
```
